Matplotlib for machine learning

Useful for making plots and graphs

```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plots
import numpy as np
x = np.linspace(0,10,100) # linespace means equally spaced values between 0 to 100
x
```

```python
y = np.sin(x)
print(y)
```

```python
z = np.cos(x)
print(z)
```

```python
# plotting the data
# sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding title x-axis and y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
# parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'r+')
```

```python
plt.plot(x,y, 'g.')
```

```python
plt.plot(x,y,'rx')
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
```

```python
plt.plot(x,np.cos(x),'r--')
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

Bar PLot

```python
fig = plt.figure()# this is empty figure so, adding axes to it
ax = fig.add_axes([0,0,1,1])# here 0,0 will be axes and 1 is height another 1 is for width
languages = ['English','French','Spanish','Latin','Germany']
people = [100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

Pie chart

```python
fig = plt.figure()# this is empty figure so, adding axes to it
ax = fig.add_axes([0,0,1,1])# here 0,0 will be axes and 1 is height another 1 is for width
languages = ['English','French','Spanish','Latin','Germany']
people = [100,50,150,40,70]
ax.pie(people,labels=languages,autopct = '%1.1f%%')
plt.show()
```

Scatter plot

Mostly useful in clustering applications

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color = 'g')
ax.scatter(x,z,color = 'b')
plt.show()
```

3D Scatter plot

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
